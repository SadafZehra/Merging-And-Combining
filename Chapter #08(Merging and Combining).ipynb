```python
 df7 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],'data1': range(7)})
df8= pd.DataFrame({'rkey': ['a', 'b', 'd'],'data2': range(3)})
pd.merge(df7, df8, left_on='lkey', right_on='rkey') 
```

```python
pd.merge(df7, df8, left_on='lkey', right_on='rkey',how="left") 
```

```python
pd.merge(df7, df8, left_on='lkey', right_on='rkey',how="right") 
```

```python
pd.merge(df7, df8, left_on='lkey', right_on='rkey',how="outer") 
```

```python
left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],index=['a', 'c', 'e'],columns=['Ohio', 'Nevada'])
right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],index=['b', 'c', 'd', 'e'],
                      columns=['Missouri', 'Alabama'])
```

```python
left2
```

```python
right2
```

```python
left2.join(right2)
```

```python
left2.join(right2,how="outer")
```

```python
left2.join(right2,how="right")
```

```python
left2.join(right2,how="inner")
```
